```python
import matplotlib.pyplot as plt
import numpy as np

from ipywidgets import interact

# constants
a = b = c = d = 1

def observe(result, state):
    result.append(state)

def update(state, Dt):
    x, y, t = state
    x = x + (a*x-b*x*y)*Dt
    y = y + (-c*y+d*x*y)*Dt
    t = t + Dt
    return x, y, t
```

```python
def simulate(Dt=0.01):
    state = [0.1, 0.1, 0.]
    result = [state]

    while state[-1] < 50.:
        state = update(state, Dt)
        observe(result, state)
    
    return result

@interact(Dt=(0.0001, 0.02, 0.001))
def simulate_interact(Dt=0.0001):
    result = simulate(Dt)

    xs, ys, timestamps = np.transpose(result)
    plt.plot(timestamps, xs)
    plt.plot(timestamps, ys)
    plt.ylabel('amount')
    plt.xlabel('time')
    plt.show()
```

```python
result = simulate()

xs, ys, timestamps = np.transpose(result)
plt.plot(xs, ys)
plt.ylabel('x')
plt.xlabel('y')
plt.show()
```
