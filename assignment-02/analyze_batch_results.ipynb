```python
import pandas as pd

batch_results = pd.read_csv("971007d_2k_steps.csv")
```

```python
batch_results.head()
```

```python
max_fire_by_condition = batch_results.groupby(["p_tree", "p_sprout", "p_propagate", "p_lightning"]).agg({"burn_prop": "max"}).reset_index()
```

```python
import seaborn as sns

sns.scatterplot(x = "p_tree", y = "burn_prop", hue = "p_lightning", data = max_fire_by_condition)
```

```python
sns.violinplot(x = "p_tree", y = "burn_prop", hue = "p_propagate",data = max_fire_by_condition)
```
