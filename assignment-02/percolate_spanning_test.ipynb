```python
from percolation import *
from core_simulation import (forestSimulation, ForestConfig)
from post_stats import *
import numpy as np

density = .30

config = ForestConfig(width=10, height=10, p_tree=density,
                    p_sprout=0, p_propagate=0, p_lightning=0, seed=17)

sim = forestSimulation(config)

results = sim.simulate(steps=250)
full_history = sim.history
```
